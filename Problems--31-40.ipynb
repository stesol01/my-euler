In England the currency is made up of pound, £, and pence, p, and there are eight coins in general circulation:

1p, 2p, 5p, 10p, 20p, 50p, £1 (100p) and £2 (200p).

It is possible to make £2 in the following way:

1×£1 + 1×50p + 2×20p + 1×5p + 1×2p + 3×1p

How many different ways can £2 be made using any number of coins?

```python
coins = {"1p":1, "2p":2, "5p":5, "10":10, "20p":20, "50p":50, "100p":1, "200p":200}

target = 200
coins = [1, 2, 5, 10, 20, 50, 100, 200]
ways = [1] + [0]*target

for coin in coins:
    for i in range(coin, target+1):
        ways[i] += ways[i-coin]

print("Ways to make change =", ways[target])
```

145 is a curious number, as 1! + 4! + 5! = 1 + 24 + 120 = 145.

Find the sum of all numbers which are equal to the sum of the factorial of their digits.

Note: as 1! = 1 and 2! = 2 are not sums they are not included.

```python
def digitsInNumber(nbr):
    digits = []
    while nbr > 0:
        digits.insert(0, nbr%10)
        nbr //= 10
    return digits

memonic = dict()
def factorial(nbr):
    if nbr in memonic:
        return memonic[nbr]
    factor = 1
    for n in range(1, nbr+1):
        factor *= n
    memonic[nbr] = factor
    return factor

def p34():
    curiousNumber = []
    for n in range(3, 1000000):
        sumOfDigits = 0
        for digit in digitsInNumber(n):
            sumOfDigits += factorial(digit)
        if n == sumOfDigits:
            curiousNumber.append(n)
    sumOfCuriousNumbers = 0
    for n in curiousNumber:
        sumOfCuriousNumbers += n
    return sumOfCuriousNumbers
    
p34()
```

The number, 197, is called a circular prime because all rotations of the digits: 197, 971, and 719, are themselves prime.

There are thirteen such primes below 100: 2, 3, 5, 7, 11, 13, 17, 31, 37, 71, 73, 79, and 97.

How many circular primes are there below one million?

```python
def digitsInNumber(nbr):
    digits = []
    while nbr > 0:
        digits.insert(0, nbr%10)
        nbr //= 10
    return digits

def getNumber(digits):
    nbr = 0
    for n in range(len(digits)):
        nbr += digits[len(digits)-n-1] * 10**n
    return nbr

def isPrime(nbr):
    if(nbr == 1): return False
    if(nbr == 2): return True
    if(nbr == 3): return True
    if(nbr == 5): return True
    if(nbr % 2 == 0): return False
    if(nbr % 3 == 0): return False
    if(nbr % 5 == 0): return False
    limit = int(nbr**0.5 + 1)
    for n in range(3,limit,2):
        if(nbr % n == 0):
            return False
    return True

def circularGenerator(nbr):
    digits = digitsInNumber(nbr)
    length = len(digits)
    for m in range(length-1):
        new = []
        for n in range(1,len(digits)):
            new.append(digits[n])
        new.append(digits[0])
        digits = new
        yield digits
        

def allPerms(elements):
    if len(elements) <=1:
        yield elements
    else:
        for perm in allPerms(elements[1:]):
            for i in range(len(elements)):
                # nb elements[0:1] works in both string and list contexts
                yield perm[:i] + elements[0:1] + perm[i:]
                  
def p35():
    circularPrimes = []
    for n in range(1000000):
        if isPrime(n) and not(n in circularPrimes):
            circular = True
            for perm in circularGenerator(n):
                nbr = getNumber(perm)
                if not(isPrime(nbr)):
                    circular = False
                    break
            if circular:
                circularPrimes.append(n)
    return circularPrimes

print(len(p35()))
```

Find the sum of all numbers, less than one million, which are palindromic in base 10 and base 2.

(Please note that the palindromic number, in either base, may not include leading zeros.)

```python
def testPalindrome(nbr):
    strNbr = str(nbr)
    strBin = str(bin(nbr)[2:])
    strNbrRev = strNbr[::-1]
    strBinRev = strBin[::-1]
    if strBinRev[len(strBinRev)-1] == 0:
        return False
    for n in range(0, len(strNbr)):
        if(strNbr[n] != strNbrRev[n]):
            return False
    for n in range(0, len(strBin)):
        if(strBin[n] != strBinRev[n]):
            return False
    return True

def p36():
    sumOfPalindromes = 0
    for n in range(1000000):
        if testPalindrome(n):
            sumOfPalindromes += n
    print(sumOfPalindromes)

p36()
```

The number 3797 has an interesting property. Being prime itself, it is possible to continuously remove digits from left to right, and remain prime at each stage: 3797, 797, 97, and 7. Similarly we can work from right to left: 3797, 379, 37, and 3.

Find the sum of the only eleven primes that are both truncatable from left to right and right to left.

NOTE: 2, 3, 5, and 7 are not considered to be truncatable primes.

```python
def isPrime(nbr):
    if(nbr == 1): return False
    if(nbr == 2): return True
    if(nbr == 3): return True
    if(nbr == 5): return True
    if(nbr % 2 == 0): return False
    if(nbr % 3 == 0): return False
    if(nbr % 5 == 0): return False
    limit = int(nbr**0.5 + 1)
    for n in range(3,limit,2):
        if(nbr % n == 0):
            return False
    return True

def p37():
    sumOfTruncatablePrimes = 0
    for n in range(11,1000000):
        if isPrime(n):
            strNbr = str(n)
            i = 0
            truncatable = True
            while i < len(strNbr)-1:
                left = strNbr[1+i:len(strNbr)]
                right = strNbr[:len(strNbr)-i-1]
                
                if not(isPrime(int(left))):
                    truncatable = False
                    break
                if not(isPrime(int(right))):
                    truncatable = False
                    break
                #print(n, left, right)
                i += 1
            if truncatable:
                sumOfTruncatablePrimes += n
    return sumOfTruncatablePrimes
                
p37()
```

If p is the perimeter of a right angle triangle with integer length sides, {a,b,c}, there are exactly three solutions for p = 120.

{20,48,52}, {24,45,51}, {30,40,50}

For which value of p ≤ 1000, is the number of solutions maximised?

```python
%%time

iterations = 0
solutions = dict()
for p in range(4, 1000, 2):
    solutions[p] = 0
    for a in range(1, p):
        for b in range(a, p-a):
            c = p - a - b
            iterations += 1
            #print("a:", a, " b:", b, " c:", c, " p:", p, " iterations:", iterations)
            if (a**2 + b**2 == c**2):
                solutions[p] += 1
pmax = max(solutions, key=lambda p: solutions[p])
print("pmax:", pmax, " nbr of sols:", solutions[pmax])
```
